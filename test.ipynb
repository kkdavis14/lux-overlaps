```python
from luxy import PeopleGroups
```

```python
pg = PeopleGroups().filter(name="tolkien").get()

print(pg.num_results)
print(pg.view_url)
```

```python
pages = pg.get_page_data_all(start_page=0, end_page=3)
print(pages)
```

```python
pages = list(pages)
```

```python
page1_items = pg.get_items(pages[0])
print(pg.get_item_data(page1_items[0])["_label"])
```

```python
for i, page in enumerate(pg.get_page_data_all(start_page=0, end_page=3), 1):
    print(f"Page {i}:", page["id"])
    for j, item in enumerate(pg.get_items(page)):
        item_data = pg.get_item_data(item)
        birth_year = None
        death_year = None
        if 'born' in item_data and 'timespan' in item_data['born']:
            birth_ts = item_data['born']['timespan']
            if 'identified_by' in birth_ts and birth_ts['identified_by']:
                birth_year = birth_ts['identified_by'][0]['content'][:4]
        if 'died' in item_data and 'timespan' in item_data['died']:
            death_ts = item_data['died']['timespan'] 
            if 'identified_by' in death_ts and death_ts['identified_by']:
                death_year = death_ts['identified_by'][0]['content'][:4]
        years = f" ({birth_year}-{death_year})" if birth_year or death_year else ""
        print(f"Item {j}:", item_data["_label"] + years)
```
